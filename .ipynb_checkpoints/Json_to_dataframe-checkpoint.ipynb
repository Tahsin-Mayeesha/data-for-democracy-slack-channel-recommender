# Convert json data to csv. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
channels  = pd.read_json('https://raw.githubusercontent.com/Data4Democracy/micro-projects/master/d4d-slack-ecosystem/channels20170617.json')
```

```python
channels.head()
```

```python
channels['channels'][0]
```

```python
from pandas.io.json import json_normalize
```

```python
channel_df = json_normalize(channels['channels'])
```

```python
channel_df.to_csv("slack_channels_d4d.csv", index = False, encoding = 'utf-8')
```
